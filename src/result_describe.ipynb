```python
res = load_json("result_valid.json")
```

```python
distri = []
```

```python
for i, name in enumerate(res):
    identities = res[name]
    print('{} 同名作者共{}个'.format(name,len(identities)))
    for j, papers in enumerate(identities):
        print('作者{}：论文{}篇'.format(j, len(papers)))
        distri.append(len(papers))
        if(len(papers) > 100):
            print('名为{}的identities[{}]: 论文{}篇'.format(name, j, len(papers)))
            
```
